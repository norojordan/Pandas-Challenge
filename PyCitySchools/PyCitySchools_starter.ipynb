```python
# Dependencies and Setup
import pandas as pd

# Files to Load 
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# Calculate the total number of schools in the school district. 
# Calculate the total number of students in the school district.

school_count = len(school_data_complete["school_name"].unique())
student_count = len(school_data_complete["Student ID"])
```

```python
# Calculate the total budget for the district. 

total_budget = school_data["budget"].sum()
#print(total_budget)

# Format with dollars, commas and round off
#currency_budget = "${:,.2f}".format(total_budget)
#print(currency_budget)
```

```python
# Calculate the average math score for all the students in the district.
# Calculate the average reading score for all the students in the district.

avg_math = school_data_complete["math_score"].mean()
avg_reading = school_data_complete["reading_score"].mean()
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
# Calculate the percentage of students with a passing reading score (70 or greater)
# Calculate the percentage of students who passed BOTH math and reading.

passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
percent_math_passing = (passing_math_count)/(student_count)*100

passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
percent_reading_passing = (passing_reading_count)/(student_count)*100

overall_passing_count = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]["student_name"].count()
percent_overall_passing = (overall_passing_count)/(student_count)*100



#print(passing_math_count)
#print(passing_reading_count)
#print(overall_passing_count)
#print(percent_math_passing)
#print(percent_reading_passing)
#print(percent_overall_passing) 

                                                                                                    
```

```python
#Create a Data Frame to hold the results.

district_summary_df = pd.DataFrame({"Total Schools": [school_count],
                                    "Total Students": [student_count],
                                    "Total Budget": [total_budget],
                                    "Average Math Score": [avg_math], 
                                    "Average Reading Score": [avg_reading], 
                                    "% Passing Math": [percent_math_passing],
                                    "% Passing Reading": [percent_reading_passing], 
                                    "% Overall Passing": [percent_overall_passing]})


#Format
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

district_summary_df
```

```python
# Use GroupBy in order to separate the data into key metrics for each school.

# Index using the name of each school.

grouped_school_type = school_data.set_index("school_name")["type"]
each_school_count = school_data_complete["school_name"].value_counts()

# Calculate Budgets and format after using in formulas
grouped_school_budget = school_data.groupby(["school_name"]).mean()["budget"]
budget_per_student = grouped_school_budget/each_school_count

# Calculate Average Test Scores
grouped_avg_math_score = school_data_complete.groupby(["school_name"])["math_score"].mean()
grouped_avg_reading_score = school_data_complete.groupby(["school_name"])["reading_score"].mean()

# Calculate number of students in each school with passing math scores and find the percentage
school_passing_math = school_data_complete[(school_data_complete["math_score"] >= 70)]
grouped_passing_math_count = school_passing_math.groupby(["school_name"])["student_name"].count()
grouped_percent_math_pass = (grouped_passing_math_count)/(each_school_count)*100

# Calculate number of students in each school with passing reading scores and find the percentage
school_passing_reading = school_data_complete[(school_data_complete["reading_score"] >= 70)]
grouped_passing_reading_count = school_passing_reading.groupby(["school_name"])["student_name"].count()
grouped_percent_reading_pass = (grouped_passing_reading_count)/(each_school_count)*100


# Calculate number of students in each school with passing reading AND math scores and find the percentage                                                               
school_passing_both = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]
grouped_passing_both_count = school_passing_both.groupby(["school_name"])["student_name"].count()
grouped_percent_both_pass = (grouped_passing_both_count)/(each_school_count)*100


# Create a data frame to test results.
grouped_schools_summary = pd.DataFrame({"School Type": grouped_school_type,
                                        "Total Students": each_school_count,
                                        "Total School Budget": grouped_school_budget,
                                         "Per Student Budget": budget_per_student,
                                         "Average Math Score": grouped_avg_math_score,
                                         "Averge Reading Score": grouped_avg_reading_score,
                                         "% Passing Math": grouped_percent_math_pass,
                                         "% Passing Reading": grouped_percent_reading_pass,
                                         "% Passing Overall": grouped_percent_both_pass})

# Format budget numbers with currency
#grouped_schools_summary["Total School Budget"] = grouped_schools_summary["Total School Budget"].map("${:,.2f}".format)
#grouped_schools_summary["Per Student Budget"] = grouped_schools_summary["Per Student Budget"].map("${:,.2f}".format)

grouped_schools_summary
```

## Math Scores by Grade

```python

# Create a data series of scores for each grade using a conditional
frosh = school_data_complete[(school_data_complete["grade"] == "9th")]
soph = school_data_complete[(school_data_complete["grade"] == "10th")]
junior = school_data_complete[(school_data_complete["grade"] == "11th")]
senior = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group each by school name and find the average
frosh_scores = frosh.groupby(["school_name"])["math_score"].mean()
soph_scores = soph.groupby(["school_name"])["math_score"].mean()
junior_scores = junior.groupby(["school_name"])["math_score"].mean()
senior_scores = senior.groupby(["school_name"])["math_score"].mean()

# Combine the series into a dataframe
math_scores_summary = pd.DataFrame({"9th": frosh_scores, "10th": soph_scores, "11th": junior_scores, "12th":senior_scores})


# Format the data. How do I adjust the spacing of the output? Come back to this if time
math_scores_summary = math_scores_summary[["9th" , "10th" , "11th" , "12th"]]
math_scores_summary.index.name = None

math_scores_summary

```
